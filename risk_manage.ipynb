```python
import pypdf

def extract_text_from_pdf(pdf_path):
    pdf_reader = pypdf.PdfReader(pdf_path)
    text = ""
    for page in pdf_reader.pages:
        text += page.extract_text() + "\n"
    return text

# Example usage
pdf_path = "/content/Netflix-10-K-01272025.pdf"  # Replace with your actual file
report_text = extract_text_from_pdf(pdf_path)
```

```python
def analyze_risk(report_text):
    prompt = f"""
          You are a financial analyst specializing in risk management.

          Analyze the following 10-K financial report and extract key financial, operational, and regulatory risks:

          {report_text[:3000]} """ # Limiting input size

    inputs = tokenizer(prompt, return_tensors="pt")
    outputs = model.generate(**inputs, max_new_tokens=500)

    return tokenizer.decode(outputs[0], skip_special_tokens=True)

# Run analysis
risk_analysis = analyze_risk(report_text)
print(risk_analysis)
```
